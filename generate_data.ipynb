# Vorschlag für das Monatsabschlussporjekt

```python
import csv
import random
```

```python
apothekenProdukte =  ["Paracetamol", "Ibuprofen", "Ibuprofen800", "Cetirizin", "Aspirin", "Iberogast", "Nasenspray", "Hustenbonbon", "Amoxicillin", "Ramipril"]
preisliste = {}
for p in apothekenProdukte:
    preisliste[p] = round(random.uniform(1.0, 10.0), 2)
print(preisliste)
```

```python
# Datensatz mit zufälligen käufen der Produkte simulieren
daten_liste = []

for i in range(300):
    zufall_produkt = random.choice(apothekenProdukte)
    zufall_menge = random.randint(1, 5)
    
    einzel_preis = preisliste[zufall_produkt]
    
    gesamtpreis = round(zufall_menge * einzel_preis, 2)
    
    # Eintrag für die CSV erstellen. Die Bibliothek braucht das Format dictionary
    verkauf = {
        "Kauf_ID": i + 1,
        "Produkt": zufall_produkt,
        "Menge": zufall_menge,
        "Einzelpreis": einzel_preis,
        "Gesamtpreis": gesamtpreis
    }
    daten_liste.append(verkauf)
```

```python
# Datensatz erweitern -> Mengenrabatt einführen
# simuliere, dass die Apotheke an diesem Verkaufstag eine Aktion hat, bei der ab einem Einkaufswert von 20 Euro die Kunden 10% Rabatt erhalten

def berechne_rabatt(verkauf_dict):
    gesamt = verkauf_dict["Menge"] * verkauf_dict["Einzelpreis"]
    
    if gesamt >= 20:
        verkauf_dict["Rabatt"] = "10%"
        verkauf_dict["BezahlterPreis"] = round(gesamt * 0.9, 2)
    else:
        verkauf_dict["Rabatt"] = "0%"
        verkauf_dict["BezahlterPreis"] = round(gesamt, 2)

    verschreibungspflichtig = ["Amoxicillin", "Ramipril", "Ibuprofen800"]

    if (verkauf_dict["Produkt"] in verschreibungspflichtig):
        verkauf_dict["verschreibungspflichtig"] = "Ja"
    else:
        verkauf_dict["verschreibungspflichtig"] = "Nein"

# auf alle Daten anwenden
for v in daten_liste:
    berechne_rabatt(v)
```

```python
spalten = list(daten_liste[0].keys())
spalten
```

```python
dateiname = "apotheken_verkaeufe.csv"

# note: excel benutzt "," als Dezimaltrennzeichen, python benutzt ".".
# um damit umzugehen, beim Datenladen Datentyperkennung ausschalten und bei PowerQuery manuell machen
# Typ ändern -> mit Gebietsschema -> Englisch (USA)
with open(dateiname, mode="w", newline="", encoding="utf-8") as datei:
    writer = csv.DictWriter(datei, fieldnames=spalten)
    writer.writeheader()
    writer.writerows(daten_liste)
```
